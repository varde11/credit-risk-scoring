1) **Remplacer les NaN de NumberOfDpendents par 0** vue la proportion faible de NaN, cela n'affecterait pas tellement le modèle.

Le bins ne semble pas être une option à cause des faibles corr, il serait préférable de faire un modèle de regression sur MonthlyIncome après avoir traiter les valeurs de NumberOfDependents.

   La première solution semble donc être la mieux adaptée, en plus en regardant les quantiles de NumberOfDependents, on voit que celui à 75% n'est qu'à 1.

Nous allons à présent commencer les prédictions sur SeriousDlqin2yrs en utilisant le RandomForestClassifier de sklearn.ensemble, mais nous avons une classe qui domine sur l'autre, la classe 0 avec 109396 données qui domine la classe 1 avec 8162 données. 

Pour régler le problème, trois approches peuvent être utilisées.

D'abord, on utilisera le paramètre class_weight='balanced' de notre modèle pour équilibrer les classe. Ensuite, on appliquera une méthode de resampling avec la fonction resample de sklearn.utils pour gonfler la classe minoritaire avec des exemples déjà existants. Enfin, on utilisera une autre méthode de resampling avec la fonction SMOTE de imblearn.over_sampling pour gonfler la classe minoritaire avec des exemples générés en fonction des premières observations.
